```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn import linear_model

x = np.linspace(1, 5, 10)
x
```

```python
l=[]
for n in x:
    temp=(np.random.choice(range(1,3)))
    l.append(n+temp)
y=np.array(l)
y
    
```

```python
plt.scatter(x,y)
plt.show()
```

```python
import numpy as np
SumX = np.sum(x)
SumY = np.sum(y)
X2Sum = np.sum(x * x)
SumXY = np.sum(x * y)
SumX2=np.sum(x)**2
```

```python
b=(len(x)*SumXY-SumX*SumY)/(len(x)*X2Sum-SumX2)    //Slope
b
```

```python
//Intercept
a=(SumY-(b*SumX))/len(x)  
a
```

```python
Yp=a+b*x
Yp
```

```python
plt.scatter(x,y)
plt.plot(x,Yp,color='Green')
plt.legend(['Predicted value','Original value'])
```
